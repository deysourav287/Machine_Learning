converting the labels into numeric form either 0 or 1

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

label encoding of breast dataset

```python
cancer_data= pd.read_csv('/content/breast_cancer_data.csv')
cancer_data.head()
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder fn
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 represents benign
1 represents malignant

```python
cancer_data['target'].value_counts()
```
